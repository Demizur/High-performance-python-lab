```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np
```

# Logistic map:
$x_{n+1} = r*x_n*(1-x_n)$

```python
#define logistic map equation
def logistic_map(r, x):
    return r * x * (1 - x)
```

```python
#plot the evolution of x
r = 3
N = 30
x = np.zeros(N)
x[0] = 0.04

for n in range(1,N):
    x[n] = logistic_map(r, x[n-1])

plt.figure(figsize=(16, 16))
plt.plot(x,'o-')
plt.title(f'Evolution of x for $ x_0 = {x[0]} ,r = {r} $', fontsize = 16)
plt.tick_params(axis='both', which='major', labelsize=20)
plt.xlabel('n', fontsize = 16)
plt.ylabel('x', fontsize = 16)
```

```python

#plot the evolution of x
N = 30
x = np.zeros(N)

fig = plt.figure(figsize=(16, 16))

def animate(i):
     # clear axes object
    plt.cla()

    r = 0  # initial value
    x[0] = 0.04
    for _ in range(i + 1): # animated iterations
        r += 0.1
        for n in range(1,N):
            x[n] = logistic_map(r, x[n-1]) 
       
    plt.title(f'Evolution of x for $ x_0 = {x[0]} ,r = {r} $', fontsize = 16)
    plt.tick_params(axis='both', which='major', labelsize=20)
    plt.xlabel('n', fontsize = 16)
    plt.ylabel('x', fontsize = 16)
    plt.ylim([0, 1])

    l = plt.plot(x,'o-')
    return l

# call the animator	 
anim = FuncAnimation(fig, animate, frames=40, interval=525, blit=True)

# save the animation as a gif file 
anim.save('biff.gif')
```

```python
# save m last x
def SaveX(r,x0,m):
    N = 1000
    x = np.zeros(N)
    x[0] = x0

    for n in range(1,N):
        x[n] = logistic_map(r, x[n-1])
    return x[-m:]
```

$x^{{(r)}}_{{n}} = r * x^{{(r)}}_{{n-1}} * (1 - x^{{(r)}}_{{n-1}})$

```python
m=200
N = 1000
r = np.linspace(0, 4.0, N)
x = np.zeros((len(r),m))
x0 = 0.03

#plot bifurcation map
plt.figure(figsize=(16, 16))
plt.title(f"Bifurcation Map for $ x_0 ={x0}, n = {N}$", fontsize = 16)
plt.tick_params(axis='both', which='major', labelsize=20)
plt.xlabel('r', fontsize=16)
plt.ylabel('$x^{(r)}_{n}$', fontsize=16)

for i,r_i in enumerate(r):
    x[i,:] = SaveX(r_i,x0,m)
    plt.plot(r_i * np.ones(m), x[i,:], ',k')
```
